#Lets clean up this data

```
# Helper functions for HTML clean-up.

class HTMLStripper(HTMLParser):
    def __init__(self):
        super().__init__() #Just Learned about what super does
        self.reset()
        self.fed = []
    def handle_data(self, d):
        self.fed.append(d)
    def get_data(self):
        return ' '.join(self.fed)

def strip_tags(html):
    s = HTMLStripper()
    s.feed(html)
    return s.get_data()
```

```
#Clear up some of the HTML in the comments
df['text']  = df['text'].apply(html.unescape)
df['text'] = df['text'].apply(strip_tags)
df.head(10)
```

#Let Lord Vader determine the sentiment of these comments...

```
vadanalyzer = SentimentIntensityAnalyzer()
```

```
def vaderize(sentence):
    return vadanalyzer.polarity_scores(sentence)
```

```
# Get vader scores and add to df
df['scores'] = df['text'].apply(vaderize)
df[['vad_neg', 'vad_neu', 'vad_pos', 'compound']] = df.scores.apply(pd.Series)
df = df.drop(columns="scores")
df.head(10)
```

```
for text in df.sort_values(by='vad_neg', ascending=False)['text'].head(20):
    print(text, '\n')
```

```
def get_toxicity(comment):
    data_dict = {
        'comment': {'text': comment},
        'languages': ['en'],
        'requestedAttributes': {'TOXICITY': {}}
    }
    response = requests.post(url=url, data=json.dumps(data_dict)) 
    response_dict = json.loads(response.content)
    try:
        return response_dict['attributeScores']['TOXICITY']['summaryScore']['value']
    except:
        return np.nan
```
